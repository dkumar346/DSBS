```python
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug'],
    'performance': [80, 75, 90, 85, 95, 70, 88, 65]
}

df = pd.DataFrame(data)

# Group by region and find the month with the highest and lowest performance
def best_worst_performance(group):
  best_month = group['month'].loc[group['performance'].idxmax()]
  worst_month = group['month'].loc[group['performance'].idxmin()]
  best_performance = group['performance'].max()
  worst_performance = group['performance'].min()
  return pd.Series({
      'Best Month': best_month,
      'Worst Month': worst_month,
      'Best Performance': best_performance,
      'Worst Performance': worst_performance
  })

# Apply the function to each region group
result = df.groupby('region').apply(best_worst_performance)

# Print the result (consider formatting for better readability)
print(result)
```
